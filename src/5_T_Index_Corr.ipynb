```python
def corr_(x1, x2):
    result = spearmanr(x1, x2)
    return result[0], result[1]
```

```python
mt_li=[]
at_li=[]
mcsa_li=[]
acsa_li=[]

mp_csa_li=[]
ms_csa_li=[]

mp_csa_ai_li=[]
ms_csa_ai_li=[]

for f in df['연번(No.)']:
    a1=df1[df1['No']==f]
    a2=df2[df2['No']==f]
    a3=df3[df3['No']==f]
    a4=df4[df4['No']==f]
    #33print(f)
    #print(a1['mp_sn'].item(),a2['mp_sn'].item(),a3['mp_sn'].item(),a4['mp_sn'].item())
    #print(a1['ms_sn'].item(),a2['ms_sn'].item(),a3['ms_sn'].item(),a4['ms_sn'].item())
    mt=np.nansum([np.nanmean([a1['mp_sn'].item(),a2['mp_sn'].item(),a3['mp_sn'].item(),a4['mp_sn'].item()]),
                  np.nanmean([a1['ms_sn'].item(),a2['ms_sn'].item(),a3['ms_sn'].item(),a4['ms_sn'].item()])])
    at=np.nansum([np.nanmean([a1['mp_sn_ai'].item(),a2['mp_sn_ai'].item(),a3['mp_sn_ai'].item(),a4['mp_sn_ai'].item()]),
                  np.nanmean([a1['ms_sn_ai'].item(),a2['ms_sn_ai'].item(),a3['ms_sn_ai'].item(),a4['ms_sn_ai'].item()])])
    mcsa=np.nansum([np.nanmean([a1['mp_csa'].item(),a2['mp_csa'].item(),a3['mp_csa'].item(),a4['mp_csa'].item()]),
                    np.nanmean([a1['ms_csa'].item(),a2['ms_csa'].item(),a3['ms_csa'].item(),a4['ms_csa'].item()])])
    acsa=np.nansum([np.nanmean([a1['mp_csa_ai'].item(),a2['mp_csa_ai'].item(),a3['mp_csa_ai'].item(),a4['mp_csa_ai'].item()]),
                    np.nanmean([a1['ms_csa_ai'].item(),a2['ms_csa_ai'].item(),a3['ms_csa_ai'].item(),a4['ms_csa_ai'].item()])])
    mt_li.append(mt)
    at_li.append(at)
    mcsa_li.append(mcsa)
    acsa_li.append(acsa)
    
    mp_csa=np.nanmean([a1['mp_csa'].item(),a2['mp_csa'].item(),a3['mp_csa'].item(),a4['mp_csa'].item()])
    ms_csa=np.nanmean([a1['ms_csa'].item(),a2['ms_csa'].item(),a3['ms_csa'].item(),a4['ms_csa'].item()])
    
    mp_csa_ai=np.nanmean([a1['mp_csa_ai'].item(),a2['mp_csa_ai'].item(),a3['mp_csa_ai'].item(),a4['mp_csa_ai'].item()])
    ms_csa_ai=np.nanmean([a1['ms_csa_ai'].item(),a2['ms_csa_ai'].item(),a3['ms_csa_ai'].item(),a4['ms_csa_ai'].item()])
    
    
    mp_csa_li.append(mp_csa)
    ms_csa_li.append(ms_csa)
    mp_csa_ai_li.append(mp_csa_ai)
    ms_csa_ai_li.append(ms_csa_ai)
```

```python
# Data   Load
excel_path = 'T_Index_Extract.csv'
df = pd.read_csv(excel_path)

# Drop nan data
df = df.dropna(subset=['m_sn'])

# define factors
# 1. TLS(Total Length of Stay)
dis_date=df['Discharge_date'].values
op_date=df['Op_date'].values
TLS = [(datetime.datetime.strptime(dis, "%Y-%m-%d") - datetime.datetime.strptime(op, "%Y-%m-%d")).days for dis, op in zip(dis_date, op_date)]
# 2. WIT
wit = df['WIT'].values.astype(float)
# 3. Cr-diff
cr_dif = np.array(df['Postop_Cr_1D']) - np.array(df['Preop_cr']).astype(float)
# 4. EBL
ebl = df['EBL'].fillna(0.).values.astype(float)
# 5. Op Time
op_time = df['Total Op time'].astype(float)

# Compare method
# 1. PADUA
padua_score = df['PADUA_Score'].values.astype(float)
# 2. RENAL
renal_score = df['RENAL_Score'].values.astype(float)
# 3. C index
c_index_score_x = df['x_C_index'].values.astype(float)
c_index_score_y = df['y_C_index'].values.astype(float)
c_index_score = np.sqrt(c_index_score_x**2 + c_index_score_y**2)
# 4. m_sn
m_sn = df['m_sn'].values.astype(float)
# 5. Tindex (mp_sn + ms_sn)
mps_sn = (df['mp_sn']+df['ms_sn']).values.astype(float)
```

```python
z, p_value = corr_(wit, padua_score)
print('z : ', z, ', p_value : ', p_value)
```
